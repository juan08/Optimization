# Step 1: Data reading from Excel

```python
excel = pd.ExcelFile('DataSet1.xlsx')
FixCost = pd.read_excel(excel,'Fix Cost')
MatrixCost = pd.read_excel(excel,'Matrix Cost')
Demand = pd.read_excel(excel,'DemandNode')
```

## Creating different sets for our problem

```python
nodos = [k for k in range(len(Demand))]
nodos
```

```python
arc = [(i,j) for i in nodos for j in nodos]
arc
```

### Initializing differents sets

```python
#Initializing demand of each node
demand={k: row['Demand'] for k, row in Demand.iterrows()}
demand
```

```python
#Initializing cost of building plan in a giving node
fixcost={k: row['Cost'] for k, row in FixCost.iterrows()}
fixcost
```

```python
#Initializing distance matrix 
c=[]
for i in nodos:
    for name in MatrixCost.columns[1:]:
        c.append(MatrixCost[name][i])
        
d = {arc[k]: c[k] for k in range(len(c))}
d
```

# Step 2: Building the model in CPLEX

```python
from docplex.mp.model import Model
```

```python
mdl=Model('F_Location1')
```

### Variable declarations

```python
x=mdl.binary_var_dict(nodos,name = 'x')
```

```python
y=mdl.integer_var_dict(arc,name = 'y')
```

### Objective function

```python
#for k in nodos:
mdl.minimize(mdl.sum(x[i]*fixcost[i] for i in nodos)+0.01*mdl.sum(y[(k,j)]*d[(k,j)]*demand[k] for k in nodos for j in nodos))
```

### Restrictions

```python
#Arcos
for k in nodos:
    mdl.add_constraint(mdl.sum(y[(k,j)] for j in nodos)==1)
```

```python
#Arcos
for k in nodos:
    for j in nodos:
        mdl.add_constraint(y[(k,j)] <= x[j])
```

```python
print(mdl.export_to_string())
```

```python
solucion = mdl.solve(log_output=True)
```
