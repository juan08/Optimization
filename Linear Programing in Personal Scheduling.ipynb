```python
#Here we apply the desagregation of demand according to the day of the week
salesHour=[]
for i in range(len(sales['Sales'])):
    if (sales['Week Day'][i]=='domingo' or sales['Week Day'][i]=='sábado'):
        salesHour.append(sales['Sales'][i]*Weekeend)
    else:
        salesHour.append(sales['Sales'][i]*DayWeek)

print(type(salesHour))
len(salesHour)
```

```python
#IN THIS PART, WE CALCULATE THE DEMAND OF PERSONNEL PER HOURS
import numpy as np
import math # Library used to round up numbers

empleRequirement=[]
for i in range(len(salesHour)): #se corta tamanio de m
    empleRequirement.append([math.ceil(num/1000) for num in salesHour[i]])

empleRequirement

```

# Step 2: Building the model in CPLEX

### Initializing differents sets

```python
Numshift = [i for i in range(len(Shift['Shift']))] #Num of Workshift 
Numhour = [i for i in range(len(Shift.columns[1:]))] #Num of hour in the workshift
workShiftNode = [(i,j) for i in Numshift for j in Numhour] # defining the numer of shift and the hours

#Initialization of shift matrix
s=[]
for i in Numshift:
    for j in Numhour:
        s.append(Shift.iloc[i][j+1])
                 
workShift = {workShiftNode[k]: s[k] for k in range(len(s))}

#Initialization of cost matrix
costShift=[CostShift['Shift Cost'][i] for i in Numshift]

```

### decraring the decision variable

```python
solArray = [] 
objArra = []
for c in range(len(empleRequirement)):  

    from docplex.mp.model import Model
    mdl=Model('PersonalScheduling')
    X=mdl.integer_var_dict(Numshift,name = 'x')
    
        #Objective function
    mdl.minimize(mdl.sum(X[i]*costShift[i] for i in Numshift))

    #demand must be satisfied
    for j in Numhour:
        mdl.add_constraint(mdl.sum(X[k]*workShift[(k,j)] for k in Numshift)>=empleRequirement[c][j])

    #full time workship must be assigned  
    #for i in range(7):
    #    mdl.add_constraint(X[i]>=1)
    
    #at minimun one full tourn must be assigned
    mdl.add_constraint(mdl.sum(X[i] for i in range(7))>=1)
        
    #solve the model    
    solucion=mdl.solve(log_output=False)
    
 
    solArray.append({i: int(X[i].solution_value) for i in Numshift if X[i].solution_value > 0.9})
    objArra.append(solucion.objective_value)
    
    
```

```python
#Objective function values
print("Cost of planing per day: ")
print(objArra)
print("\n")

#Total cost for the planning
print('Total planing cost = ',sum(objArra))
```

# Step 3: PL assign the work shift to employees

### Definig the sets for the variables

```python
#Num of day for the horizontal planning
numDays = len(sales['Week Day']) 
```

```python
#listing the set for the weekends
setWeekend = [i for i in range(len(sales['Week Day'])) if (sales['Week Day'][i]=='domingo' or sales['Week Day'][i]=='sábado')]
setWeekend
```

```python
#listing the umber of employee
numEmployee = [i for i in range(len(Availability['Employee']))]
numEmployee
```

```python
#listing the number of workshift
Numshift = [i for i in range(len(Shift['Shift']))]
Numshift
```

```python
#defining the set for all the possible altenative in the DV (employee, workshift, day)
s = []
for k in range(numDays):
    s.append([(i,j,k) for i in numEmployee for j in solArray[k]])

availaSet = []
for i in range(len(s)):
    for j in range(len(s[i])):
        availaSet.append(s[i][j])
#print(availaSet)
```

```python
#create matrix for employee availability
#Here we read the worker disponibility
e=[]
for i in numEmployee:
    for j in Numshift:
        e.append(Availability.iloc[i][j+1])

availability = np.asarray(e,dtype= int) #Transform the list into array type  
availability = np.reshape(availability,(len(numEmployee),len(Numshift)))
print(availability)
```

### Building the LP model for the assigning scheduling

```python
from docplex.mp.model import Model
mdl2=Model('AssigPersonal')
```

```python
#create desicion variable
Y=mdl2.binary_var_dict(availaSet,name = 'Y')
```

```python
#Objective function
mdl2.maximize(mdl.sum(Y[i] for i in availaSet))
```

```python
#The required numer of perssonal must be assined to the workshift in a specific day
for k in range(numDays): 
    for j in solArray[k]:
        mdl2.add_constraint(mdl2.sum(Y[(i,j,k)] for i in numEmployee)==solArray[k][j])
```

```python
#One employee must be assigned one time to a work shift in a particular day
for k in range(numDays): 
    for i in numEmployee:
        mdl2.add_constraint(mdl2.sum(Y[(i,j,k)] for j in solArray[k])<=1)
```

```python
#every assigned workshift must respect the employee disponibility
for k in range(numDays): 
    for i in numEmployee:
        mdl2.add_constraints(Y[(i,j,k)]<=100*availability[i][j] for j in solArray[k])
```

```python
#the employee must work maximun half the 
for i in numEmployee:
    mdl2.add_constraint(mdl2.sum(Y[(i,j,k)] for k in setWeekend for j in solArray[k])<=5)
```

# PART 3: Output result

```python
#Generating planning per day
SM = [] # monthly scheduling
dayWork = pd.DataFrame()
for k in range(numDays):
    schedulDay = []
    for i in numEmployee:
        for j in solArray[k]:
            schedulDay.append(finalSchedule[0][(i,j,k)])

    schedulDay = np.asarray(schedulDay,dtype= int) #Transform the list into array type  
    schedulDay = np.reshape(schedulDay,(len(numEmployee),len(solArray[k]))) #reshaping the number to a matrix
    
    schedulDay = pd.DataFrame(schedulDay) #transforming to a DF
    schedulDay.columns = [i for i in solArray[k]] #changing the names of the colums
    
    SM.append(schedulDay)

#Generating report for scheduling per month

empDay = pd.DataFrame(SM[0].sum(axis=1))
empDay.columns = [str(0)] 
for i in range(1,numDays):
    empDayPlus = pd.DataFrame(SM[i].sum(axis=1))
    empDayPlus.columns = [str(i)] 
    empDay=empDay.join(empDayPlus)
    
#Printing the repord
with pd.ExcelWriter('employeeSchedule.xlsx') as writer: 
    empDay.to_excel(writer, sheet_name='Month planing') #Monthly planning
    for k in range(len(SM)):
        SM[k].to_excel(writer, sheet_name=str(k)) #day by day planning

#for i in range(numDays)
#    print(SM[i].sum(axis=1))
    
```
